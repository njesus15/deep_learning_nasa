```python
import pickle 
import numpy as np
from keras.models import load_model
from keras import Model
import matplotlib.pyplot as plt
import sys
sys.path.insert(0, '/Users/jesusnavarro/Desktop/trail_project/trainer/data/')
from DataSet import DataSet
import random
from tensorflow.python.lib.io import file_io
from imageio import imread
import keras
import numpy as np
from imageio import imread
from keras import Model
from keras.layers import Conv2D
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers import MaxPooling2D
from keras.models import Sequential
from keras.models import load_model
# Questions:
# How were the convolution parameters chosen?
from tensorflow.python.lib.io import file_io


def create_trail_model(mean=0.0, std=0.1, s=None):
    k = keras.initializers.RandomNormal(mean=mean, stddev=std, seed=s)
    # Initialising the CNN
    classifier = Sequential()

    # Step 1 - Convolution
    classifier.add(Conv2D(32, (3, 3), input_shape=(101, 101, 3), activation='relu', kernel_initializer=k))

    # Step 2 - Pooling
    classifier.add(MaxPooling2D(pool_size=(2, 2)))

    # Adding a second convolutional layer
    classifier.add(Conv2D(32, (3, 3), activation='relu', kernel_initializer=k))
    classifier.add(MaxPooling2D(pool_size=(2, 2)))

    # Adding a third convolutional layer
    classifier.add(Conv2D(32, (3, 3), activation='relu', kernel_initializer=k))
    classifier.add(MaxPooling2D(pool_size=(2, 2)))

    # Adding a fourth convolutional layer
    classifier.add(Conv2D(32, (3, 3), activation='relu', kernel_initializer=k))
    classifier.add(MaxPooling2D(pool_size=(2, 2)))

    classifier.add(Conv2D(32, (3, 3), activation='relu', kernel_initializer=k))
    classifier.add(MaxPooling2D(pool_size=(2, 2)))


    # Step 4 - Full connection
    classifier.add(Dense(units=4096, activation='relu', kernel_initializer=k))
    classifier.add(Dense(units=3, activation='sigmoid'))

    # Compiling the CNN
    classifier.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    return classifier


def rep_layer_ouptut(paths, model, logs=True):
    """
    Computes the prediction of the neural network 'model'.
    In this application, the prediction of 'model' is the representational layer and not the softmax activation.

    :param paths: List of tuples containing (img_path_names, class label)
    :param model: Keras model without output layer

    :return: (ndarray) images, (int) number of images
    """
    output = []
    size = len(paths)
    i = 1
    labels = []
    KEY = {'right': np.array([1, 0, 0]),
           'center': np.array([0, 1, 0]),
           'left': np.array([0, 0, 1])}

    for file in paths:
        img_path = file[0]
        label = file[1]
        labels.append(KEY[label])
        # get img array
        img_arr, _ = read_img_io(img_path)
        int_out = model.predict(img_arr)
        output.append(int_out)
        if size > 100 and logs:
            if i % float((size // 100)) == 0.0:  # logs of loaded data percentage
                print("Output % {:g} of the data".format(float('{:.4g}'.format(100.0 * float(i) / float(size)))))
            i += 1
    return np.array(output).reshape(size, -1), np.array(labels)

def read_img_io(file_path_df):
    """
    Helper method to read data into array

    :param file_path_df: data frame containing file name paths and class lables
    :return: numpy array of image representation and numpy array of labels

    """
    key = {'right': np.array([1, 0, 0]),
           'center': np.array([0, 1, 0]),
           'left': np.array([0, 0, 1])}

    if not isinstance(file_path_df, list):  # Nest as a list if itsa singleton
        file_path = [file_path_df]
    else:
        file_path = file_path_df

    labels = []
    imgs = []

    for paths in file_path:
        if len(paths) == 2:  # images, labels
            file_name = paths[0]
            class_ = key[paths[1]]
            labels.append(class_)
        else:
            file_name = paths  # dataframe only contains path string

        with file_io.FileIO(file_name, 'rb') as f:
            img = f.read()
            img = imread(img)
            img = img * (1.0 / 255.0)
            imgs.append(img)

    return np.array(imgs), np.array(labels)

```

# KNN Performance on Trained Neural Nets

```python
from sklearn.neighbors import KNeighborsClassifier

root = '/Users/jesusnavarro/Desktop/trail_forest_results/results/trained_models/'

models = ['deepnn_subset_001.h5', 
          'deepnn_subset_full_exc_001.h5',
          'deepnn_subset_001_and_002.h5',
          'deepnn_subset_003.h5']

keras_models = [load_model(root + model_name) for model_name in models]

data_001 = DataSet(root_dir='/Users/jesusnavarro/Desktop/DataSet/',
                   train_subsets=[],
                   test_subsets=['/001/'],
                   loc='local')

data_df = data_001.test_set[2]
random.shuffle(data_df)
train_df = data_df[:-2000]
test_df = data_df[-2000:]


```

```python
accuracy = []
weight_tensors = []


for model in keras_models:
    rep_layer = Model(inputs=model.input, outputs=model.get_layer(index=7).output)
    kfit = KNeighborsClassifier()
    x_train, y_train = rep_layer_ouptut(train_df, rep_layer)
    x_test, y_test = rep_layer_ouptut(test_df, rep_layer)
    
    kfit.fit(x_train, y_train)
    acc = kfit.score(x_test, y_test)
    accuracy.append(acc)
    weight_tensors.append(rep_layer.get_weights())
```

```python
plt.subplots(1,1, figsize=(10,6))
plt.bar(['1 = ' + str(accuracy[0]),
         '*1 = ' + str(accuracy[1]),
         '1&2 = ' + str(accuracy[2]),
         '3 = ' + str(accuracy[3])], accuracy)
plt.title('KNN Performance on Trained NN on data 001')
print(accuracy)
```

# Single Dense Layer KNN Performance


```python
import torch.nn
from sklearn.utils import shuffle

input_size = 3 * 101*101
data_001 = DataSet(root_dir='/Users/jesusnavarro/Desktop/DataSet/',
                   train_subsets=[],
                   test_subsets=['/001/'],
                   type='arr'
                   )
datax, datay = data_001.test_set[0], data_001.test_set[1]

datax, datay = shuffle(datax, datay)

train_x, train_y = datax[:-2000], datay[:-2000]
test_x, test_y = datax[-2000:], datay[-2000:]


train_x = torch.tensor(train_x)
train_x = train_x.view(train_x.size(0), input_size)

test_x = torch.tensor(test_x)
test_x = test_x.view(test_x.size(0), input_size)
```

```python
dim = [2048, 1024, 512, 256, 128, 64, 32, 4]

for d in dim:
    
    dense = torch.nn.Sequential(torch.nn.Linear(input_size, d),
                                torch.nn.ReLU()).float()

    out_train = dense(train_x.float())
    out_test = dense(test_x.float())

    clf = KNeighborsClassifier()
    clf.fit(out_train.detach().numpy(), train_y)
    s = clf.score(out_test.detach().numpy(), test_y)
    print("Accuracy with single dense layer (RELU) dim " + str(d) + " :", s)
```

```python

for d in dim:
    
    dense = torch.nn.Sequential(torch.nn.Linear(input_size, d)).float()

    out_train = dense(train_x.float())
    out_test = dense(test_x.float())

    clf = KNeighborsClassifier()
    clf.fit(out_train.detach().numpy(), train_y)
    s = clf.score(out_test.detach().numpy(), test_y)
    print("Accuracy with single dense layer dim " + str(d) + " :", s)
```

```python
def create_trail_model_uni(minv= -0.2, maxv= 0.2, s=None):
    k = keras.initializers.RandomUniform(minval=minv, maxval=maxv, seed=s)
    # Initialising the CNN
    classifier = Sequential()

    # Step 1 - Convolution
    classifier.add(Conv2D(32, (3, 3), input_shape=(101, 101, 3), activation='relu', kernel_initializer=k))

    # Step 2 - Pooling
    classifier.add(MaxPooling2D(pool_size=(2, 2)))

    # Adding a second convolutional layer
    classifier.add(Conv2D(32, (3, 3), activation='relu', kernel_initializer=k))
    classifier.add(MaxPooling2D(pool_size=(2, 2)))

    # Adding a third convolutional layer
    classifier.add(Conv2D(32, (3, 3), activation='relu', kernel_initializer=k))
    classifier.add(MaxPooling2D(pool_size=(2, 2)))

    # Adding a fourth convolutional layer
    classifier.add(Conv2D(32, (3, 3), activation='relu', kernel_initializer=k))
    classifier.add(MaxPooling2D(pool_size=(2, 2)))

    classifier.add(Conv2D(32, (3, 3), activation='relu', kernel_initializer=k))
    classifier.add(MaxPooling2D(pool_size=(2, 2)))


    # Step 4 - Full connection
    classifier.add(Dense(units=4096, activation='relu', kernel_initializer=k))
    classifier.add(Dense(units=3, activation='sigmoid'))

    # Compiling the CNN
    classifier.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    return classifier
```
